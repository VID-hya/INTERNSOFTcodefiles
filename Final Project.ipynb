```python
# initiating pyttsx3 engine with speech systhesizers
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty("voice",'voices[0].id')
```

```python
# defining a funtion to speak a sentence
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
# defining the function to greet the person
def wishMe():
    hour = datetime.datetime.now().hour
    if hour>=0 and hour<=12:
        speak("hello, good mornig Ms.Vidhya.Hope you have a great day ahead")
        print("hello, Good mornig Ms.Vidhya.Hope you have a great day ahead")
    elif hour>=12 and hour<=19:
        speak("hello, good afternoon Ms.Vidhya")
        print("hello, Good Afternoon Ms.Vidhya")
    else:
        speak("It's getting late and you had a long and an eventfull day.Get some rest Ms.Vidhya")
        print("It's getting late and you had a long and an eventfull day.Get some rest Ms.Vidhya")
```

```python
def takeCommand():
    spch = sr.Recognizer()
    with sr.Microphone() as source:
        print("I am listening...")
        audio = spch.listen(source)
        
        try:
            statement = spch.recognize_google(audio,language='en-in')
            print(f"user said:{statement}\n")
            
        except Exception as e:
            speak("pardon me, please say that again")
            return "None"
        return statement
            
       
```
